```python
### Multiple plots

## sample data

x=[1,2,3,4,5]
y1=[1,4,9,16,25]
y2=[1,2,3,4,5]

plt.figure(figsize=(9,5))
plt.subplot(1,2,1)
plt.plot(x,y1,color='red',linestyle='--',marker='*',linewidth=2,markersize=15)
plt.subplot(1,2,2)
plt.plot(x,y2,color='red',linestyle='--',marker='*',linewidth=2,markersize=15)

```

```python
### Multiple plots

## sample data

x=[1,2,3,4,5]
y1=[1,4,9,16,25]
y2=[1,2,3,4,5]

plt.figure(figsize=(9,5))
plt.subplot(1,2,1)
plt.plot(x,y1,color='green',linestyle='-',marker='*',linewidth=2,markersize=15)
plt.title("1")

plt.subplot(1,2,2)
plt.plot(x,y1,color='red',linestyle='--',marker='*',linewidth=2,markersize=15)
plt.title("2")

plt.subplot(2,2,1)
plt.plot(x,y2,color='blue',linestyle='--',marker='*',linewidth=2,markersize=15)
plt.title("3")

plt.subplot(2,2,2)
plt.plot(x,y2,color='red',linestyle='--',marker='*',linewidth=2,markersize=15)
plt.title("4")
```

```python
import matplotlib.pyplot as plt

# Sample data
x = [1, 2, 3, 4, 5]
y1 = [1, 4, 9, 16, 25]
y2 = [1, 2, 3, 4, 5]

plt.figure(figsize=(10, 8))

# Subplot 1
plt.subplot(2, 2, 1)
plt.plot(x, y1, color='green', linestyle='--', marker='*', linewidth=2, markersize=15)
plt.title("1")

# Subplot 2
plt.subplot(2, 2, 2)
plt.plot(x, y1, color='red', linestyle=':', marker='*', linewidth=2, markersize=15)
plt.title("2")

# Subplot 3
plt.subplot(2, 2, 3)
plt.plot(x, y2, color='blue', linestyle='--', marker='*', linewidth=2, markersize=15)
plt.title("3")

# Subplot 4
plt.subplot(2, 2, 4)
plt.plot(x, y2, color='red', linestyle='--', marker='*', linewidth=2, markersize=15)
plt.title("4")

# plt.tight_layout()  # Adjust spacing to prevent overlap
# plt.show()
```

```python
import seaborn as sns
tips=sns.load_dataset('tips')
```

```python
sns.scatterplot(x='total_bill',y='tip',data=tips)
plt.title('Scatter plot using Seaborn')
```

```python
## Line plot
sns.lineplot(x='size',y='total_bill',data=tips)
```

```python
sns.barplot(x='size',y='total_bill',data=tips)
```

```python
sns.boxplot(x='size',y='total_bill',data=tips)
```

```python
sns.violinplot(x='size',y='total_bill',data=tips)
```

```python
sns.histplot(tips['total_bill'],bins=10,kde=True)
```

```python
### KDEplot

sns.kdeplot(tips['total_bill'],fill=True)
```
