```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
tips=sns.load_dataset("tips")
flights=sns.load_dataset("flights")
```

```python
tips.tail()
```

```python
#displot is used for showing only one variable(univariate)
#kde stands for kernal density estimation. The guassian line is called as kde(kernal density estimate)
sns.distplot(tips['total_bill'],kde=True,bins=10)
plt.ylabel("count")
```

```python
#jointplot is used to compare two varibales in a data here we compare total bill in x axis tip in y axis
#jointplots displays a relationship between 2 varibales (bivariate)
sns.jointplot(x="total_bill",y="tip",data=tips)
```

# Categorical Data

```python
#A barplot is basically used to aggregate the categorical data according to some methods and by default its the mean. It can also be understood as a visualization of the group by action. To use this plot we choose a categorical column for the x-axis and a numerical column for the y-axis, and we see that it creates a plot taking a mean per categorical column.
sns.barplot(x="sex",y="total_bill",data=tips)
```

```python
import numpy as np
sns.barplot(x="sex",y="total_bill",data=tips,estimator=np.std)
```

```python
#countplot() seaborn. countplot() method is used to Show the counts of observations in each categorical bin using bars. we should take ony categorical data in x axis. It result the count of the data using group by 
sns.countplot(x="sex",data=tips)
```

# Grids

```python
iris=sns.load_dataset("iris")
```

```python
iris
```

```python
iris['species'].unique()
```

```python
#This object maps each variable in a dataset onto a column and row in a grid of multiple axes. Different axes-level plotting functions can be used to draw bivariate plots in the upper and lower triangles, and the the marginal distribution of each variable can be shown on the diagonal.


g=sns.PairGrid(iris)
g.map(plt.scatter)
```

```python
g=sns.PairGrid(iris)
g.map_diag(sns.distplot)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
```

```python
# maps a dataset onto multiple axes arrayed in a grid of rows and columns that correspond to levels of variables in the dataset. The plots it produces are often called lattice, trellis, or small-multiple graphics.
g=sns.FacetGrid(data=tips,col='time',row='smoker')
g.map(sns.distplot,'total_bill')
```

```python
g=sns.FacetGrid(data=tips,col='time',row='smoker')
g.map(plt.scatter,'total_bill','tip')
```
