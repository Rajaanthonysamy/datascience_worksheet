```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
```

```python
x=np.linspace(0,5,11)
y=x**2
```

```python
x,y
```

```python
plt.plot(x,y,'-r')
plt.xlabel("X lable")
plt.ylabel("Y lable")
plt.title("Title")
```

```python
#in screen how much map we want to print   SUBPLOT
plt.subplot(1,2,1)
plt.plot(x,y,'r')
plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
#oops method  

fig=plt.figure()
axes=fig.add_axes([0.1,0.1,0.8,0.8])
axes2=fig.add_axes([0.2,0.5,0.4,0.3])
axes.plot(x,y)
axes2.plot(y,x)
```

```python
fig,axes=plt.subplots(nrows=1,ncols=2)
axes[0].plot(x,y)
axes[1].plot(y,x)
```

```python
#fixing size of figure and DPI
fig=plt.figure(figsize=(8,5))
ax=fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
#for subplots
fig,axes=plt.subplots(nrows=2,ncols=1,figsize=(8,8))
axes[0].plot(x,y)
axes[1].plot(x,y)
```

```python
fig=plt.figure()

ax=fig.add_axes([0,0,1,1])
ax.plot(x,y)
ax.set_title('Title')
ax.set_ylabel('Y')
ax.set_xlabel("X")

```

```python
fig.savefig('referencefig.png',dpi=200)
```

```python
#legends
fig=plt.figure()

ax=fig.add_axes([0,0,1,1])
ax.plot(x,x**2,label="X squared")
ax.plot(x,x**3,label="X cuubed")
ax.legend(loc=0)
```

# Plot appearance

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,y,color='red',linewidth=1,linestyle='--',alpha=0.5,marker='o',markersize=20,markerfacecolor="yellow",
        markeredgewidth=5,markeredgecolor="green")
#ax.plot(x,y,color='#FF8c00') RGB hexcode
# linewidth is sizeof line
# alpha is used to visibliy scale
#linestyle is for specify the style like dot line r dashline
```

# Control over axis appearance

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,x**3,color='black',lw='2',ls='--',marker="o")
ax.set_xlim([0,1])
ax.set_ylim([0,1])
```
